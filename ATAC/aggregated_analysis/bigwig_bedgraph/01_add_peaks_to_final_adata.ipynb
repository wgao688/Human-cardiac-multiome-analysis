### Update the cell type annotation and filter the fragment adata to the barcodes in the final ATAC adata

```python
shared_barcodes = set(fragment_adata.obs_names) & set(final_ATAC_adata.obs_names)
shared_barcodes = list(shared_barcodes)
len(shared_barcodes)
```

```python
filt_fragment_adata = fragment_adata[shared_barcodes, :]
```

```python
filt_fragment_adata.obs['cell_type'] = final_ATAC_adata.obs['cell_type']
```

```python
%%time
filt_fragment_adata.write("final_adata_with_fragments.h5ad")
```
