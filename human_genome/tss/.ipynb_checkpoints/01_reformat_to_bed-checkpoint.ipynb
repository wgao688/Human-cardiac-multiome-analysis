### Need to reformat the downloaded tss to a bed format

```python
import pandas as pd
```

```python
ensembl_tss = pd.read_csv("tss.txt", delim_whitespace=True, header=None)
ensembl_tss.columns = ['chr', 'ENSEMBL', 'TSS', 'strand', 'no', 'ensembl_id']
ensembl_tss.head()
```

```python
ensembl_tss[['ensembl_symbol', 'gene_symbol']] = ensembl_tss['ensembl_id'].str.split('-', expand=True).iloc[:, 0:2]
ensembl_tss = ensembl_tss.drop(columns = ['ensembl_id'])
ensembl_tss.head()
```

```python
chromosome_df = pd.read_csv("chromosomes.txt", delim_whitespace=True, header=None)
chromosome_df.columns = ['chr', 'symbol_chr']
chromosome_df['symbol_chr'] = "chr" + chromosome_df['symbol_chr']
```

```python
joined_df = ensembl_tss.merge(chromosome_df, on = "chr")
```

```python
bed_df = joined_df[["symbol_chr", "TSS", "TSS", "gene_symbol"]]
bed_df.head()
```

```python
# save in bed format
bed_df.to_csv("tss.bed", header=None, sep = "\t", index=None)
```
